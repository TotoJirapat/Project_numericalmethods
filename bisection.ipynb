```python
import math
x = 0.2
true_value = math.cos(x)

#cos(x) = 1 - x^2/2! + x^4/4!
t1 = 1
t2 = -x**2 / math.factorial(2)
t3 = x**4 / math.factorial(4)

approx_value = t1 + t2 + t3

tolerance = abs(true_value - approx_value)

print(f"cos(0.2) = {approx_value}")
print(f"cos(0.2) = {true_value}")
print(f"tolerance = {tolerance}")
```

```python
import math
x = 0.3
true_value = math.exp(-x)

#e^(-x) = 1 - x + x^2/2! - x^3/3!
t1 = 1
t2 = -x
t3 = x**2 / math.factorial(2)
t4 = -x**3 / math.factorial(3)

approx_value = t1 + t2 + t3 + t4
tolerance = abs(true_value - approx_value)

print(f"e^(-0.3) = {approx_value}")
print(f"e^(-0.3) = {true_value}")
print(f"tolerance = {tolerance}")
```

```python
import math

x = 0.1

t1 = x
t2 = -x**2 / 2

approx_ln = t1 + t2

true_ln = math.log(1.1)

tolerance = abs(true_ln - approx_ln)


print(f"ln(1.1) = {approx_ln}")
print(f"ln(1.1) = {true_ln}")
print(f"tolerance = {tolerance}")
```
