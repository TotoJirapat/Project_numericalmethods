```python
import math
```

```python
def fixed_point(g,x0,epsilon ,N):
    x = x0
    for k in range(N):
        xk = g(x)
        if abs(x-xk) < epsilon:
            return xk
        x = xk
    print("The iteration did not converge after ",N," iterations")
```

```python
def g(x):
    return math.exp(-x)
```

```python
x0 = 0.5
epsilon = 0.005
N = 50
fixed_point(g, x0, epsilon, N)
```

```python
x0 = 1.5      
epsilon = 1e-6
N = 100
fixed_point(g, x0, epsilon, N)
```

```python
import math

def g(x):
    return math.log(math.log(x + 4) + 2)

def fixed_point(g,x0,epsilon ,N):
    x = x0
    for k in range(1, N + 1):
        x_new = g(x)
        error = abs(x_new - x)
        
        print(f"{k:<3} {x:<12.8f} {x_new:<12.8f} {error:<12.8f}")
        
        if error < epsilon:
            print(f"\nConverged to x = {x_new:.8f} in {k} iterations.")
            return x_new
        
        x = x_new
    print("The iteration did not converge after ",N," iterations")

# เริ่มคำนวณ
x0 = 1.2
epsilon = 1e-6
N = 100

fixed_point(g, x0, epsilon, N)
```
