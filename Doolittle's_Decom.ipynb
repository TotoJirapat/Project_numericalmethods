```python
import numpy as np
def doolittle_lu(A):
    A = np.array(A, dtype=float)
    n = A.shape[0]

    if A.shape[0] != A.shape[1]:
        raise ValueError("A ต้องเป็นเมทริกซ์จัตุรัส")

    L = np.eye(n)
    U = np.zeros((n, n), dtype=float)

    for i in range(n):
        for k in range(i, n):
            U[i][k] = A[i][k] - sum(L[i][j] * U[j][k] for j in range(i))

        if abs(U[i][i]) < 1e-12:
            raise ZeroDivisionError(f"Pivot U[{i},{i}] = 0 ทำให้หารศูนย์ไม่ได้")

        for k in range(i+1, n):
            L[k][i] = (A[k][i] - sum(L[k][j] * U[j][i] for j in range(i))) / U[i][i]

    return L, U
```

```python
def forward_substitution(L, b):
    n = len(b)
    y = np.zeros_like(b, dtype=float)
    for i in range(n):
        sum_Ly = 0
        for j in range(i):
            sum_Ly += L[i][j] * y[j]
        if abs(L[i][i]) < 1e-12:
            raise ZeroDivisionError(f"L[{i},{i}] = 0 ทำให้หารศูนย์ไม่ได้")
        y[i] = (b[i] - sum_Ly) / L[i][i]
    return y
```

```python
def back_substitution(U, y):
    n = len(y)
    z = np.zeros_like(y, dtype=float)
    for i in range(n-1, -1, -1):
        sum_Uz = 0
        for j in range(i+1, n):
            sum_Uz += U[i][j] * z[j]
        if abs(U[i][i]) < 1e-12:
            raise ZeroDivisionError(f"U[{i},{i}] = 0 ทำให้หารศูนย์ไม่ได้")
        z[i] = (y[i] - sum_Uz) / U[i][i]
    return z
```

```python
def doolittle_solver(A, b):
    try:
        L, U = doolittle_lu(A)
        y = forward_substitution(L, b)
        z = back_substitution(U, y)
        return z
    except Exception as e:
        return f"ไม่สามารถแก้สมการได้: {e}"
```

```python
A = np.array([
    [2, -1, 1],
    [4, 1, -1],
    [-2, 2, 3]
], dtype=float)
b = np.array([2, 7, 3], dtype=float)

z = doolittle_solver(A, b)
print(z)
```
