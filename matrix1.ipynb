```python
def forward_substitution_augmented(A):
    n = len(A)
    x = [0] * n

    for i in range(n):
        sum_Lx = 0
        for j in range(i):
            sum_Lx += A[i][j] * x[j]
        x[i] = (A[i][-1] - sum_Lx) / A[i][i]

    return x
```
