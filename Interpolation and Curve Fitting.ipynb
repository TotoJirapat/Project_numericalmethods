```python
import math
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([50, 70, 80, 100, 120, 150, 180, 200]).reshape(-1, 1)
Y = np.array([1.8, 2.4, 2.5, 3.1, 3.4, 4.0, 4.8, 5.2])

model = LinearRegression()
model.fit(X, Y)
```

```python
slope = model.coef_[0]        
intercept = model.intercept_

print(f"β1 (slope): {slope:.4f}")
print(f"β0 (intercept): {intercept:.4f}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, color='blue', label='Actual data')

plt.plot(X, model.predict(X), color='red', label='Regression Line')

plt.xlabel('Land Size (sq.wah)')
plt.ylabel('Selling Price (million baht)')
plt.title('Relationship Between Land Size and Selling Price')

plt.legend()
plt.grid(True)

plt.show()
```

```python
area_to_predict = np.array([[135]])
predicted_price = model.predict(area_to_predict)[0]

print(f"Estimated price for 135 sq.wah: {predicted_price:.2f} million baht")
```

```python
def lagrange_interpolation(x_values, y_values, x):
    n = len(x_values)
    p = 0
    for j in range(n):
        L = 1
        for i in range(n):
            if i != j:
                L *= (x - x_values[i]) / (x_values[j] - x_values[i])
        p += y_values[j] * L
    return p
 
x_values = [0.0, 1.0, 2.0, 3.0]
y_values = [1.0, 0.0, 1.0, 2.0]
 
x_interp1 = 1.5
x_interp2 = 2.5
 
result1 = lagrange_interpolation(x_values, y_values, x_interp1)
result2 = lagrange_interpolation(x_values, y_values, x_interp2)
 
print(f"ค่าประมาณของ f({x_interp1}) = {result1:.4f}")
print(f"ค่าประมาณของ f({x_interp2}) = {result2:.4f}")
```
