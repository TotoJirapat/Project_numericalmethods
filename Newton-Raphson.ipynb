```python
import math
```

```python
def newton_raphson_method(f, df, x0, tol, N):
    k = 0
    x = x0
    while k < N:
        xk = x - f(x) / df(x)
        if abs(xk - x) < tol or abs(f(xk)) < tol:
            return xk
        x = xk
        k += 1
    raise ValueError("Newton-Raphson method did not converge within N iterations.")
```

```python
#ข้อ3
def f(x):
    return math.sin(x) - x**2 + 1

def df(x):
    return math.cos(x) - 2*x


x0 = 1.0
tol = 1e-6
N = 100


newton_raphson_method(f, df, x0, tol, N)

```

```python
#ข้อ4

def f(x):
    return x * math.log(x) - 2

def df(x):
    return math.log(x) + 1  

x0 = 2.0
tol = 1e-8
N = 100

newton_raphson_method(f, df, x0, tol, N)

```

```python
#ข้อ5
def f(x):
    return math.exp(-0.1 * x) * math.sin(x) - 0.5

def df(x):
    return math.exp(-0.1 * x) * (math.cos(x) - 0.1 * math.sin(x))

x0 = 1.0
tol = 1e-5
N = 100

newton_raphson_method(f, df, x0, tol, N)

```
