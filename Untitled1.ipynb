```python
def bisection_method(f, a, b, epsilon, N):
    n = 1
    fa = f(a)
    fb = f(b)
    if fa * fb >= 0:
        raise ValueError('f(a) and f(b) must have opposite signs')
    while n <= N:
        c = (a + b)/2
        fc = f(c)
        if abs(f(c)) < epsilon:
            return c
        if f(a) * (c) < 0:
            b = c
            fb = f(c)
        else:
            a = c
            fa = f(c)
            print(n,a,b,f(c))
            n += 1
    raise RuntimeError(f'Method failed after {N} iterations')





```

```python
def f(x):
    return math.cos(x)-x
```

```python
a = 0.0
b = 1.0
epsilon = 1e-6
N = 100
```

```python
zero = bisection_method(f, a, b, epsilon, N)
print('Approximate zero:', zero)
print('Function value at the zero',f(zero))
```
