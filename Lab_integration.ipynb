```python
def midpoint_rule(f, a, b, n):
    h = (b - a) / n
    total = 0
    for i in range(n):
        x = a + h*(i + 0.5)
        total += f(x)
    return h * total
```

```python
def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    total = f(a) + f(b)
    for i in range(1, n):
        x = a + i * h
        total += 2 * f(x)
    return h * total / 2
```

```python
def simpson_rule(f, a, b, n):
    if n % 2 == 1:
        raise ValueError("n must be even for Simpson's Rule")
    h = (b - a) / n
    total = f(a) + f(b)
    
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            total += 2 * f(x)
        else:
            total += 4 * f(x)
    return h * total / 3
```

```python
import math
def f(x):
    return math.exp(-x**2 / 2)

a = -1
b = 1

for n in [10, 100, 1000]:
    print(f"n = {n}")
    print("Midpoint:", midpoint_rule(f, a, b, n))
    print("Trapezoidal:", trapezoidal_rule(f, a, b, n))
    print("Simpson:", simpson_rule(f, a, b, n))
    print()
```

```python
import random

def monte_carlo(f, a, b, N):
    total = 0
    for _ in range(N):
        x = random.uniform(a, b)
        total += f(x)
    return (b - a) * total / N
```

```python
import math

def f(x):
    return math.exp(-x**2 / 2)

a = -1
b = 1

for n in [10, 100, 1000]:
    print(f"n = {n}")
    if n % 2 == 0:
        print("Monte Carlo:", monte_carlo(f, a, b, n))
    print()
```
