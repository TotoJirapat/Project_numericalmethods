```python
import math
def f(x):
    return math.exp(-x) - math.sin(x)

def secant_method(x0, x1, epsilon, N=50):
    iteration = 0
    ea = float('inf')
    while ea > epsilon and iteration < N:
        f_x0 = f(x0)
        f_x1 = f(x1)
        if f_x1 - f_x0 == 0:
            print("Error: Division by zero.")
            break
        x_new = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)

        if x_new != 0:
            ea = abs((x_new - x1) / x_new)
        x0, x1 = x1, x_new
        iteration += 1
    print("ค่ารากที่ประมาณได้: x =", x_new)
    print("จำนวนรอบที่ใช้:", iteration)
    print("ค่า f(x) ณ ค่ารากประมาณ =", f(x_new))
    print("ค่าความคลาดเคลื่อนสุดท้าย (|εa|) =", ea)

x0 = 0.5
x1 = 1.0
epsilon = 1e-6
N =50
secant_method(x0, x1,epsilon,N)


```

```python
def f(x):
    return x**3 - 2*x - 1  
def secant_method(x0, x1, epsilon, N):
    iteration = 1
    while iteration < N:
        f_x0 = f(x0)
        f_x1 = f(x1)

        if f_x1 - f_x0 == 0:
            print("Error: หารด้วยศูนย์")
            return None

        x2 = x0 - f_x0 * (x1 - x0) / (f_x1 - f_x0)

        if abs(f(x2)) < epsilon:
            break
        x0 = x1
        x1 = x2
        iteration += 1

    # แสดงผลลัพธ์
    print("ค่าราก (ค่า x ที่ จุดตัด) ที่ประมาณได้:", x2)
    print("จำนวนรอบที่ใช้ (iterations):", iteration)
    print("ค่า |f(x)| ทึ่ ค่ารากประมาณ:", abs(f(x2)))

x0 = 1.5
x1 = 2.0
epsilon = 1e-8
N = 100
secant_method(x0,x1,epsilon,N)
```

```python
import math

def f(x):
    return x * math.sin(x) - 1

def secant_method(x0, x1, tol=1e-7, max_iter=50):
    iteration = 0
    while iteration < max_iter:
        f_x0 = f(x0)
        f_x1 = f(x1)

        if f_x1 - f_x0 == 0:
            print("Error: Division by zero")
            return None, iteration
        
        x2 = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)
        error = abs(x2 - x1)

        if error < tol:
            return x2, iteration + 1
        
        x0, x1 = x1, x2
        iteration += 1

    print("Did not converge within max iterations.")
    return x2, iteration

initial_guesses = [
    (1.0, 1.5),
    (1.1, 1.2)
]

print("ผลลัพธ์จาก Secant Method:")
print("เริ่มต้นจาก (x0, x1) => ค่า root ที่ได้, จำนวนรอบที่ใช้")

for x0, x1 in initial_guesses:
    root, num_iter = secant_method(x0, x1)
    print(f"(x0={x0}, x1={x1}) => root ≈ {root:.10f}, รอบ = {num_iter}")
```
