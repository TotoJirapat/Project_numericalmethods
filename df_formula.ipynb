```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('timese.csv')
print(df)
```

```python
print(df.columns)
```

```python
velocity = []

for i in range(len(df)):
    if i == 0:
        # forward 
        v = (df['position(m)'][i+1] - df['position(m)'][i]) / (df['time(s)'][i+1] - df['time(s)'][i])
    elif i == len(df)-1:
        # backward 
        v = (df['position(m)'][i] - df['position(m)'][i-1]) / (df['time(s)'][i] - df['time(s)'][i-1])
    else:
        # central 
        v = (df['position(m)'][i+1] - df['position(m)'][i-1]) / (df['time(s)'][i+1] - df['time(s)'][i-1])
    velocity.append(v)

# เพิ่มคอลัมน์velocityในdataFrame
df['velocity'] = velocity


print(df)
```

```python
plt.figure(figsize=(12, 5))

#ตำแหน่ง-เวลา
plt.subplot(1, 2, 1)
plt.plot(df['time(s)'], df['position(m)'], 'o-', label='position(m)')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('position(m) vs time(s)')
plt.grid(True)
plt.legend()

# ความเร็ว-เวลา
plt.subplot(1, 2, 2)
plt.plot(df['time(s)'], df['velocity'], 's--r', label='Velocity')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs Time')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
